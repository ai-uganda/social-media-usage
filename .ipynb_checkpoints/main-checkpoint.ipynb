```python
# Change gender variable into a numeric data using pandas - to easily analyse
t = pd
Social["gender"].replace(['female', 'male'], [0, 1], inplace=True)
Social
```

```python
Social["gender"] = Social["gender"].astype(str)    # needed
Social["gender"] = Social["gender"].replace(["female", "male", np.nan], [0, 1, -1])   # worked
Social
```

When splitting - need to know what is already known
what is the input?

for this dataset we need to undestand what is already known?
- ages for 20-27, what platforms they would use and what genders they fall in



```python
# The data needs to be split so as to predict the output 

# In this case we first drop the column 'platform' which is what we are trying to predict 

X = Social.drop(columns=['platform'])
X
```

```python
# Generate the predictor output set of dataset - gives what platforms will be used by the individuals

Y = Social['platform']
Y
```

datatype of Y is object; object refers to categorical datatype

```python
# Building the model (using the Decision Tree classification method)

from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()

```

```python
# train the model using the X(input set) and Y(output set)
model.fit(X, Y)
```

```python
# make predictions using the desifned model to determine platforms used by a 21yr female and 32 yr male
prediction = model.predict([[21, 0], [32, 1]])
prediction
```

Rule of Thumb for Allocation of data sets for training and testing
Default 
- 70-80% for training model
- 20-30% for test for testing the model

Accuracy scores range from 0 - 1

- The accuracy score ranges from 0-1. The returned value shows the model is 100% accurately makes the prediction 
- However, accuracies are also affected by the test size used 

```python
# train_test_split - for testing accuracy of models
from sklearn.model_selection import train_test_split
# then we generate the accuracy scores that result from prediction
from sklearn.metrics import accuracy_score

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

model = DecisionTreeClassifier()
model.fit(X_train, Y_train)

prediction = model.predict(X_test)
 
score = accuracy_score(Y_test, prediction)
score
```

```python
# if the test size is 30% of the data 

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)

model = DecisionTreeClassifier()
model.fit(X_train, Y_train)

prediction = model.predict(X_test)

score2 = accuracy_score(Y_test, prediction)
score2
```
